```python
date_columns = ['birth_date','death_date']
#df[date_columns]=pd.to_datetime(df[date_columns],format='%Y-/%m-/%d')
df[date_columns] = pd.to_datetime(df[date_columns].stack(),format='%Y-%m-%d').unstack()
```

```python
#pd.to_datetime works on a string, list, or pd.Series. 
#dte is a pd.DataFrame and is why you are having issues. 
#dte.stack() produces a a pd.Series where all rows are stacked on top of each other. 
#However, in this stacked form, because it is a pd.Series, I can get a vectorized pd.to_datetime to work on it. 
#the subsequent unstack simply reverses the initial stack to get the original form of dte
```

```python
df['age']=(df['year']-df['birth_date'].dt.year).astype(int,errors = 'ignore')
```

```python
df['decade'] = (np.floor(df['year'] / 10) * 10).astype(int)

df['female'] = df['sex'] == 'Female'
female_winners = df.groupby('decade', as_index=False)['female'].mean()

df['male'] = df['sex'] == 'Male'
male_winners = df.groupby('decade', as_index=False)['male'].mean()

gender_prop_winners = male_winners.merge(female_winners, on='decade')
gender_prop_winners[['male','female']] = gender_prop_winners[['male','female']]*100

gender_prop_winners
```

```python
from matplotlib.pyplot import figure
figure(figsize=(10, 5))

plt.plot('decade', 'male', data=gender_prop_winners, color='gold')
plt.plot('decade', 'female', data=gender_prop_winners,color='silver')
plt.grid(True, alpha=0.3)

# Remove top and left spines:
ax = plt.gca() # Get current axis (gca)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

plt.legend()
plt.title("Gender proportion over the decades", fontsize=18)
plt.xlabel("Decade",fontsize=12)
plt.ylabel("Proportion in %",fontsize=12)
plt.text(1930,60,"Although there is a huge imbalance among the gender proportions, \na positive trend is seen among the 'female' during recent decades.",bbox=dict(facecolor='lightgreen', alpha=0.5),fontsize = 12);
```

```python
category_prop_winners = df.groupby('category', as_index=False)['female'].mean()
female_category_winners = df.groupby('category', as_index=False)['male'].mean()

category_prop_winners = category_prop_winners.merge(female_category_winners,on = 'category')
category_prop_winners[['male','female']] = round(category_prop_winners[['male','female']]*100,2)
category_prop_winners
```

```python
from matplotlib.pyplot import figure
figure(figsize=(15, 7))
plt.bar(category_prop_winners['category'], category_prop_winners['female'], alpha=0.7, width=0.4,color='gold',ec='black',label='female')
plt.bar(category_prop_winners['category'], category_prop_winners['male'], alpha=0.7, width=0.4,color='silver',ec='black',label='male')
plt.grid(True, alpha=0.3)
plt.xlabel("Category", fontsize=14)
plt.ylabel("Number of Nobel Laurets", fontsize=14)
plt.title("Gender Proportion in %", fontsize=18)
plt.legend(bbox_to_anchor=(1, 1), shadow=True, ncol=1)


values1 = category_prop_winners['female']
for index, each in enumerate(values1):
    plt.annotate(str(each)+'%', xy=(index, each), 
             xytext=(index,each+3), ha = 'center',
             bbox={'boxstyle': 'round', 'pad': 0.5, 'facecolor': 
                   'orange', 'edgecolor': 'orange', 'alpha': 0.6})
    

values2 = category_prop_winners['male']
for index, each in enumerate(values2):
    plt.annotate(str(each)+'%', xy=(index, each), 
             xytext=(index,each+3), ha = 'center',
             bbox={'boxstyle': 'round', 'pad': 0.5, 'facecolor': 
                   'orange', 'edgecolor': 'orange', 'alpha': 0.6})
    
# Remove top and left spines:
ax = plt.gca() # Get current axis (gca)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

plt.show()
```
